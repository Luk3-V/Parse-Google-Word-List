```python
import pandas as pd

pd.set_option('mode.chained_assignment', None)

words = pd.read_csv('../input/english-word-frequency/unigram_freq.csv') # Words w/ their frequency
nouns = pd.read_csv('../input/parsed-subtlex-word-list/SUBTLEXfreqPoS-nouns.csv') # Nouns w/ noun frequency
verbs = pd.read_csv('../input/parsed-subtlex-word-list/SUBTLEXfreqPoS-verbs.csv') # Verb w/ verb frequency
adjectives = pd.read_csv('../input/parsed-subtlex-word-list/SUBTLEXfreqPoS-adjectives.csv') # Adjectives w/ adjective frequency
adverbs = pd.read_csv('../input/parsed-subtlex-word-list/SUBTLEXfreqPoS-adverbs.csv') # Adverb w/ adverb frequency
prepositions = pd.read_csv('../input/parsed-subtlex-word-list/SUBTLEXfreqPoS-prepositions.csv') # Perposition w/ preposition frequency

# Rename columns
words.rename(columns={'word': 'Word', 'count': 'Web_freq'}, inplace=True)
nouns.rename(columns={'FREQcount': 'Subt_freq', 'CDcount': 'Film_freq', 'All_PoS_SUBTLEX': 'All_PoS', 'All_freqs_SUBTLEX': 'PoS_freqs'}, inplace=True)
verbs.rename(columns={'FREQcount': 'Subt_freq', 'CDcount': 'Film_freq', 'All_PoS_SUBTLEX': 'All_PoS', 'All_freqs_SUBTLEX': 'PoS_freqs'}, inplace=True)
adjectives.rename(columns={'FREQcount': 'Subt_freq', 'CDcount': 'Film_freq', 'All_PoS_SUBTLEX': 'All_PoS', 'All_freqs_SUBTLEX': 'PoS_freqs'}, inplace=True)
adverbs.rename(columns={'FREQcount': 'Subt_freq', 'CDcount': 'Film_freq', 'All_PoS_SUBTLEX': 'All_PoS', 'All_freqs_SUBTLEX': 'PoS_freqs'}, inplace=True)
prepositions.rename(columns={'FREQcount': 'Subt_freq', 'CDcount': 'Film_freq', 'All_PoS_SUBTLEX': 'All_PoS', 'All_freqs_SUBTLEX': 'PoS_freqs'}, inplace=True)

# Make words all lowercase
nouns = nouns.apply(lambda x: x.str.lower() if x.dtype=='object' else x)
verbs = verbs.apply(lambda x: x.str.lower() if x.dtype=='object' else x)
adjectives = adjectives.apply(lambda x: x.str.lower() if x.dtype=='object' else x)
adverbs = adverbs.apply(lambda x: x.str.lower() if x.dtype=='object' else x)
prepositions = prepositions.apply(lambda x: x.str.lower() if x.dtype=='object' else x)

words
```

---
# Nouns:
---

```python
# Nouns in words list
nouns2 = nouns.merge(words, on = 'Word', how = 'left')

nouns2.dropna(inplace = True)
nounsMissing = nouns[~nouns.index.isin(nouns2.index)]

nouns2.drop(columns=['FREQlow', 'Cdlow', 'SUBTLWF', 'Lg10WF', 'SUBTLCD', 'Lg10CD', 'Dom_PoS_SUBTLEX', 'Freq_dom_PoS_SUBTLEX', 'Percentage_dom_PoS'], inplace = True)

nouns2.sort_values(by=['Noun_freq', 'Web_freq', 'Word'], ascending=[False, False, True], ignore_index=True, inplace=True)

print(nounsMissing.shape)
nouns2
```

---
# Verbs:
---

```python
# Verbs in words list
verbs2 = verbs.merge(words, on = 'Word', how = 'left')

verbs2.dropna(inplace = True)
verbsMissing = verbs[~verbs.index.isin(verbs2.index)]

verbs2.drop(columns=['FREQlow', 'Cdlow', 'SUBTLWF', 'Lg10WF', 'SUBTLCD', 'Lg10CD', 'Dom_PoS_SUBTLEX', 'Freq_dom_PoS_SUBTLEX', 'Percentage_dom_PoS'], inplace = True)

verbs2.sort_values(by=['Verb_freq', 'Web_freq', 'Word'], ascending=[False, False, True], ignore_index=True, inplace=True)

print(verbsMissing.shape)
verbs2
```

---
# Adjective:
---

```python
# Adjective in words list
adjectives2 = adjectives.merge(words, on = 'Word', how = 'left')

adjectives2.dropna(inplace = True)
adjectivesMissing = adjectives[~adjectives.index.isin(adjectives2.index)]

adjectives2.drop(columns=['FREQlow', 'Cdlow', 'SUBTLWF', 'Lg10WF', 'SUBTLCD', 'Lg10CD', 'Dom_PoS_SUBTLEX', 'Freq_dom_PoS_SUBTLEX', 'Percentage_dom_PoS'], inplace = True)

adjectives2.sort_values(by=['Adjective_freq', 'Web_freq', 'Word'], ascending=[False, False, True], ignore_index=True, inplace=True)

print(adjectivesMissing.shape)
adjectives2
```

---
# Adverbs:
---

```python
# Adverbs in words list
adverbs2 = adverbs.merge(words, on = 'Word', how = 'left')

adverbs2.dropna(inplace = True)
adverbsMissing = adverbs[~adverbs.index.isin(adverbs2.index)]

adverbs2.drop(columns=['FREQlow', 'Cdlow', 'SUBTLWF', 'Lg10WF', 'SUBTLCD', 'Lg10CD', 'Dom_PoS_SUBTLEX', 'Freq_dom_PoS_SUBTLEX', 'Percentage_dom_PoS'], inplace = True)

adverbs2.sort_values(by=['Adverb_freq', 'Web_freq', 'Word'], ascending=[False, False, True], ignore_index=True, inplace=True)

print(adverbsMissing.shape)
adverbs2
```

---
# Propositions:
---

```python
# Prepositions in words list
prepositions2 = prepositions.merge(words, on = 'Word', how = 'left')

prepositions2.dropna(inplace = True)
prepositionsMissing = prepositions[~prepositions.index.isin(prepositions2.index)]

prepositions2.drop(columns=['FREQlow', 'Cdlow', 'SUBTLWF', 'Lg10WF', 'SUBTLCD', 'Lg10CD', 'Dom_PoS_SUBTLEX', 'Freq_dom_PoS_SUBTLEX', 'Percentage_dom_PoS'], inplace = True)

prepositions2.sort_values(by=['Preposition_freq', 'Web_freq', 'Word'], ascending=[False, False, True], ignore_index=True, inplace=True)

print(prepositionsMissing.shape)
prepositions2
```

---

```python
nouns2.to_csv('nouns-freq.csv',index=False)
nounsMissing.to_csv('nouns-missing.csv',index=False)
verbs2.to_csv('verbs-freq.csv',index=False)
verbsMissing.to_csv('verbs-missing.csv',index=False)
adjectives2.to_csv('adjectives-freq.csv',index=False)
adjectivesMissing.to_csv('adjectives-missing.csv',index=False)
adverbs2.to_csv('adverbs-freq.csv',index=False)
adverbsMissing.to_csv('adverbs-missing.csv',index=False)
prepositions2.to_csv('prepositions-freq.csv',index=False)
prepositionsMissing.to_csv('prepositions-missing.csv',index=False)
```
